# Óptica y Ondas — Gráficas comparativas

Este notebook contiene las dos gráficas pedidas en el trabajo:

1. **Pregunta 3:** rapidez longitudinal $v$ vs módulo de Young $Y$.
2. **Pregunta 7:** comparación entre rapidez primaria $v_p$ y rapidez secundaria $v_s$ en función de la rigidez $\mu$.

Se usan unidades consistentes:
- $Y$ y $\mu$ en **GPa** para mostrar en los ejes.
- Conversión interna a **Pa** para calcular velocidades en **m/s**.


```python
import numpy as np
import matplotlib.pyplot as plt
```

## Pregunta 3 — Rapidez $v$ vs Módulo de Young $Y$

La ecuación usada es:

$$
v = \sqrt{\frac{Y}{\rho}}
$$

con:
- $\rho = 2700\ \text{kg/m}^3$
- $Y$ en el rango $[0.01, 350]$ GPa


```python
# Parámetros para la pregunta 3
rho = 2700  # kg/m^3
Y_GPa = np.linspace(0.01, 350, 1000)
Y_Pa = Y_GPa * 1e9

# Rapidez longitudinal
v = np.sqrt(Y_Pa / rho)

# Gráfica
plt.figure(figsize=(8, 5))
plt.plot(Y_GPa, v, linewidth=2)
plt.xlabel('Módulo de Young, Y (GPa)')
plt.ylabel('Rapidez, v (m/s)')
plt.title('Pregunta 3: Rapidez longitudinal vs Módulo de Young')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Pregunta 7 — Comparación entre $v_p$ y $v_s$ vs rigidez $\mu$

Las ecuaciones usadas son:

$$
v_p = \sqrt{\frac{\lambda + 2\mu}{\rho}}
$$

$$
v_s = \sqrt{\frac{\mu}{\rho}}
$$

con:
- $\rho = 2700\ \text{kg/m}^3$
- $\lambda = 2.44\ \text{GPa}$
- $\mu$ en el rango $[0.1, 350]$ GPa


```python
# Parámetros para la pregunta 7
rho = 2700  # kg/m^3
lam_GPa = 2.44
mu_GPa = np.linspace(0.1, 350, 1000)

# Conversión a Pa
lam_Pa = lam_GPa * 1e9
mu_Pa = mu_GPa * 1e9

# Velocidades sísmicas
vp = np.sqrt((lam_Pa + 2 * mu_Pa) / rho)
vs = np.sqrt(mu_Pa / rho)

# Gráfica comparativa
plt.figure(figsize=(8, 5))
plt.plot(mu_GPa, vp, label='v_p', linewidth=2)
plt.plot(mu_GPa, vs, label='v_s', linewidth=2)
plt.xlabel('Rigidez, μ (GPa)')
plt.ylabel('Rapidez (m/s)')
plt.title('Pregunta 7: Comparación entre rapidez primaria y secundaria')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

## Conclusiones rápidas

- En la **pregunta 3**, la rapidez aumenta con el módulo de Young: a mayor rigidez, mayor velocidad de propagación.
- En la **pregunta 7**, tanto $v_p$ como $v_s$ aumentan con $\mu$, pero siempre se cumple que $v_p > v_s$.
